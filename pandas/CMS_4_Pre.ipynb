```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',200)
```

```python
people = {"first":["Sravan Reddy","Cruel","Suspect"],"last":["Rokkam","Goat","Glint"],"email":["sravanreddy.rokkam@gmail.com","cruelgoat@gmail.com","suspectglint@gmail.com"]}
```

```python
people
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
#When we compare a column of a dataframe with a static value, it returns a series of boolean values.
#This series when passed to a dataframe will return the rows satisfying the condition.
type(df['first']=='Cruel')
```

```python
#We use '|' for OR condition and '&' for AND condition when comparing between two conditions
a=(df['first']=='Cruel') | (df['last']=='Glint')
```

```python
a,type(a)
```

```python
#We can also use .loc() method and pass the filter as an argument to get the filtered rows.
#And we also have the option to fetch specific columns using .loc() method"
df.loc[a]
df.loc[~a,'email']
```

```python
df[a],type(df[a])
```
