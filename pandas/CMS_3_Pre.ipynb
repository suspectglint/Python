```python
#Indexes - set, reset, persist the index changes, sorting are covered in this notebook for a smaller dataframe.
```

```python
import pandas as pd
```

```python
people = {"first":["Sravan Reddy","Cruel","Suspect"],"last":["Rokkam","Goat","Glint"],"email":["sravanreddy.rokkam@gmail.com","cruelgoat@gmail.com","suspectglint@gmail.com"]}
```

```python
people
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df.index
```

```python
df.info()
```

```python
#When we set an index using set_index() method, it is automatically dropped from the column list.
#And also, the changes are not persisted by default. So to make it permanent, we need to enter inplace=True.
df.set_index('email',inplace=True)
```

```python
df
```

```python
df.loc['cruelgoat@gmail.com']
```

```python
df.iloc[0]
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
#And also, if we donot want the column to be dropped while we are setting it as an index, we mention drop=False.
df.set_index('email',inplace=True,drop=False)
```

```python
df
```
