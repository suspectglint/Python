```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',200)
```

```python
people = {"first":["Sravan Reddy","Cruel","Suspect"],"last":["Rokkam","Goat","Glint"],"email":["sravanreddy.rokkam@gmail.com","cruelgoat@gmail.com","suspectglint@gmail.com"]}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns
```

```python
#If we want to rename a specific column in the given dataframe, do that using df.rename() method.
df.rename(columns={'email':'email_address'},inplace=True)
```

```python
#When we pass a list of string and assign to columns, the columns in the dataframe will be renamed.
df.columns = ['first name','last name','email']
```

```python
df
```

```python
#And if we want to apply some transformation(string changes) to column names, we can do it using functions of df.columns.str 
df.columns=df.columns.str.replace(' ','_')
```

```python
df
```

```python
#Some of the other functions that are available in df.columns.str
dir(df.columns.str)
```

```python
#Now let's understand how to update rows and columns in a given dataframe.
```

```python
def transfrm(row):
    row['email']=row['email'].replace('gmail.com','hotmail.com')
```

```python
df.apply(transfrm,axis=1)
```

```python
#df.apply() - this function is used to transformtation to 
df['email'] = df['email'].apply(lambda x : x.lower())
```

```python
df
```

```python
#We can use both .at() and .loc() to update cells in a given dataframe.
#.loc is more versatile and can handle multiple rows and columns or even subsets of data.
df.loc[2,'email'] = 'sambaputra@gmail.com'
```

```python
#We can use both .at() and .loc() to update cells in a given dataframe.
#Purpose: .at is optimized for accessing or setting a single scalar value in a DataFrame or Series.
df.at[1,'email'] = 'fragilegoat@gmail.com'
```

```python
df
```

```python
df['email']=df['email'].str.lower()
```

```python
#apply() - this function is used to apply transformations at a series level in a dataframe.
df['email'].apply(len)
```

```python
#applymap() - this function is used to apply transformations to all the cells in a dataframe.
df.applymap(len)
```

```python
df
```

```python
#map() - function is used to replace/map values in a series as per the mapping shared in the input arguments.
#Now, let's try to change few of the firstnames in the dataframe
#And one thing to be noted here, it only changes the values that are mentioned in the dictionary argument. For the other values it just returned
#a NaN value(Not a Number). So to avoid such things, we can use replace() method.
df['first_name'].map({'Cruel':'Kind','Suspect':'Accept'})
```

```python
df
```

```python
#replace() - function is used to replace values in a series as per the mapping shared in the input arguments.
#Now, let's try to change few of the firstnames in the dataframe
#And one thing to be noted here, it only changes the values that are mentioned in the dictionary argument.
df['first_name'].replace({'Cruel':'Kind','Suspect':'Accept'},inplace=True)
```

```python
df['first_name']=df['first_name'].replace({'Cruel':'Kind','Suspect':'Accept'})
```
