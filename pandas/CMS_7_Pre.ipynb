```python
#In this notebook, we will look at various ways of sorting columns in pandas.
```

```python
import pandas as pd
```

```python
pokemon_data = {
    "Name": ["Pikachu", "Charmander", "Bulbasaur", "Squirtle", "Eevee"],
    "Type": ["Electric", "Fire", "Grass/Poison", "Water", "Normal"],
    "HP": [35, 39, 45, 44, 55],
    "Attack": [55, 52, 49, 48, 55],
    "Defense": [40, 43, 49, 65, 50]
}
```

```python
df =pd.DataFrame(pokemon_data)
```

```python
df
```

```python
df.sort_index(ascending=False)
```

```python
df
```

```python
df.sort_values(by='Name')
```

```python
#Here in pandas, we use .sort_values() method to sort a given dataframe based on the columns shared in the list 
#and also we can use this sort a single series of values. 
#by - to specify the column(s) on which data needs to be sorted.
#ascending - to specify the order of data sort ascending/descending for the corresponding column.
df.sort_values(by=['Attack','Name'],ascending=[False,True],inplace=True)
```

```python
df
```

```python
#This we have seen earlier in index file. This method can be used to sort the given dataframe by index.
df.sort_index(inplace=True)
```

```python
df
```

```python
#This is the scenario where we are trying to sort a single series of data.
df['Name'].sort_values(ascending=False)
```
