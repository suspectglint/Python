```python
#In this notebook, we explore the various ways to add/remove columns and rows from a dataframe.
#First, we explore how to add/remove columns followed by rows.
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',200)
```

```python
people = {"first":["Sravan","Cruel","Suspect"],"last":["Rokkam","Goat","Glint"],"email":["sravanreddy.rokkam@gmail.com","cruelgoat@gmail.com","suspectglint@gmail.com"]}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
print(df['first']+' '+df['last'])
type(df['first']+' '+df['last'])
```

```python
#Here we are adding a new column by assigning new series to a new column mentioned in square bracket notation.
df['full_name']=df['first']+' '+df['last']
```

```python
df
```

```python
#Now if we want to drop columns from a given dataframe, we can do it using .drop() method and mention the list of 
#columns to be deleted from the dataframe.
df.drop(columns=['first','last'],inplace=True)
```

```python
df
```

```python
#Here we are splitting the full_name column into first and last using str.split() method.
#And we are also passing extra attribute expand=True to make sure the values are split into separate columns. If we donot mention
#expand, it will be a single series with list of values being split.
df['full_name'].str.split(' ',expand=True)
```

```python
#Here we are assigning the split columns into two new columns
df[['first','last']]=df['full_name'].str.split(' ',expand=True)
```

```python
df
```

```python
########End of column addition and removal.##############
```

```python
#df.append() method is removed in version 2.0 and is no longer used for addition of rows. Instead we use 
#df.concat() method for adding rows to a given dataframe.
df2 = pd.DataFrame({"first":["Omar","Bunk"],
       "last":["Little","Moreland"],
       "email":["omarlittle@wire.com","bunkmoreland@@wire.com"]})
df=pd.concat([df,df2],ignore_index=True)
```

```python
df
```

```python
filter=df['last']!='Moreland'
filter
```

```python
df=df.loc[filter]
```
